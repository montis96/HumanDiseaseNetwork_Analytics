## Rimozione dei nodi di tipo Gene e degli archi collegati

```python
vertices = pd.read_csv("../Dataset/Original/diseasome [Nodes].csv")
vertices = vertices.rename(columns={'0':'kind', '1':"type"}, errors="raise")
vertices = vertices.set_index('id')
vertices
```

```python
edges = pd.read_csv("../Dataset/Original/diseasome [Edges].csv")
edges = edges.set_index('id')
edges
```

```python
to_remove=list()
for i in range(0, len(edges)):
    if vertices.loc[edges.loc[i]['Source']]['kind'] == "disease" and vertices.loc[edges.iloc[i]['Target']]['kind'] == "gene":
        to_remove.append(i)
edges = edges.drop(to_remove)
edges 
```

```python
to_remove=list()
for i in list(vertices.index.values):
    if vertices.loc[i]['kind'] == "gene":
        to_remove.append(i)
vertices = vertices.drop(to_remove)
vertices
```
